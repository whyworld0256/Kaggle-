```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model1=KNeighborsClassifier()
model2=RandomForestClassifier()
param_grid1= {'n_neighbors': range(2, 10), 'weights': ['uniform', 'distance']}
param_grid2 = {'n_estimators': range(10, 100, 10), 'criterion': ['gini', 'entropy']}

gridsearch1=GridSearchCV(model1,param_grid=param_grid1,scoring='accuracy',cv=10,refit=True)
gridsearch1.fit(best_data,y_train)
model1=gridsearch1.best_estimator_

gridsearch2=GridSearchCV(model2,param_grid=param_grid2,scoring='accuracy',cv=10,refit=True)
gridsearch2.fit(best_data,y_train)
model2=gridsearch2.best_estimator_
```
